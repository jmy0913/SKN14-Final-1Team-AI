## RAG 및 퍼플렉시티 비교

### 정성적 평가 체인

langchain 형식은 openai 무료 크레딧(하루 250만 토큰) 적용이 안돼서, open ai 형식으로 평가 코드를 수정했습니다. 근데 재시도 로직에 top_k가 많아서 이거 무료 토큰도 금방 쓰는거 같네요..

평가 기준 부분은 윤지님이 한번 더 봐주시면 감사하겠습니다. 적은 양으로 테스트해보니까 이번에 퍼플렉시티랑 차이가 거의 없어서.. 저희 rag에 유리한 기준으로 좀 바꿔야할 거 같아요..

```python
import os
import json
import pandas as pd
import numpy as np
from openai import OpenAI
from dotenv import load_dotenv

# 환경 변수 불러오기
load_dotenv()
client = OpenAI()

def api_document_evaluation(question: str, answer: str, system_answer: str):
    prompt = f"""
당신은 API 문서 기반 평가 전문가입니다. 주어진 시스템의 답변이 **기준 정답과 얼마나 일치하는지**,
**중요한 정보를 얼마나 잘 재현했는지**, 그리고 **세부 사항을 얼마나 잘 반영했는지** 평가해야 합니다.

평가 기준:
1. **정확도 (Accuracy)**:
    - 3점: 기준 정답과 거의 완벽하게 일치
    - 2점: 핵심 내용은 맞지만 일부 세부 사항이 누락되거나 다르게 표현됨
    - 1점: 주요 정보는 맞지만 중요한 요소가 크게 누락됨
    - 0점: 기준 정답과의 불일치가 큼

2. **재현율 (Recall)**:
    - 3점: 모든 핵심 정보를 정확히 재현
    - 2점: 주요 요소는 재현되었으나 일부 세부 정보가 부족
    - 1점: 핵심 정보 일부가 누락되었거나 잘못 설명됨
    - 0점: 핵심 정보가 많이 누락됨

3. **구체성 (Specificity)**:
    - 3점: 매개변수나 반환값 등을 상세하고 구체적으로 설명
    - 2점: 기본적인 설명은 포함되었으나 세부적인 설명이 부족
    - 1점: 간단한 설명은 포함되었으나, 세부 정보가 매우 부족
    - 0점: 구체적이지 않은 답변, 중요한 세부 사항이 누락됨 (핵심 정보 내용이 다르다면 구체적이어도 0점)

출력은 반드시 JSON 형식으로만 하세요:
{{
  "accuracy": <0~3>,
  "recall": <0~3>,
  "specificity": <0~3>
}}

## 사용자 질문 : {question}
## 기준 정답 : {answer}
## 시스템 답변 : {system_answer}
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        temperature=0,
        response_format={"type": "json_object"},
        messages=[
            {"role": "system", "content": "너는 API 문서 기반 RAG 답변 평가 전문가이다."},
            {"role": "user", "content": prompt},
        ]
    )

    return response.choices[0].message.content
```

```python

# === 배치 평가 실행 ===
evaluation_results = []

for idx in range(len(dataset_df)):
    question = dataset_df['user_input'][idx]
    answer = dataset_df['reference'][idx]

    # 시스템 답변 생성
    system_answer = run_langraph(question, '1', None)['answer']

    # 평가 수행 (OpenAI 호출)
    response = api_document_evaluation(question, answer, system_answer)

    # JSON 파싱
    try:
        scores = json.loads(response)
        evaluation_results.append({
            'index': idx,
            'question': question,
            'reference': answer,
            'system_answer': system_answer,
            'accuracy': scores.get('accuracy', 0),
            'recall': scores.get('recall', 0),
            'specificity': scores.get('specificity', 0)
        })
        print(f"평가 완료: {idx+1}/{len(dataset_df)} - Accuracy: {scores.get('accuracy', 0)}")
    except json.JSONDecodeError:
        print(f"JSON 파싱 오류: {idx+1}/{len(dataset_df)}")
        evaluation_results.append({
            'index': idx,
            'question': question,
            'reference': answer,
            'system_answer': system_answer,
            'accuracy': 0,
            'recall': 0,
            'specificity': 0
        })

# 결과 DataFrame 변환
results_df = pd.DataFrame(evaluation_results)

# 평균 점수 계산
avg_accuracy = results_df['accuracy'].mean()*100/3
avg_recall = results_df['recall'].mean()*100/3
avg_specificity = results_df['specificity'].mean()*100/3
overall_avg = (avg_accuracy + avg_recall + avg_specificity) / 3

print(f"\n=== 전체 평가 결과 ===")
print(f"평균 정확도: {avg_accuracy:.2f}")
print(f"평균 재현율: {avg_recall:.2f}")
print(f"평균 구체성: {avg_specificity:.2f}")
print(f"전체 평균: {overall_avg:.2f}")

# 결과 저장
results_df.to_csv('rag_batch_evaluation_results.csv', index=False)
print(f"\n결과가 'rag_batch_evaluation_results.csv'에 저장되었습니다.")
```
